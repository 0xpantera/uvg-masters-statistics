## Distribuciones multivariadas en Python

A veces podemos obtener mucha informacion de como dos variables (o mas) se relacionan si los graficamos en conjunto. Este notbeook tiene como objetivo demostrar como graficar dos variables en conjunto nos puede dar informacion que no se evidencia al graficarlas por separado.

```python
# import the packages we are going to be using
import numpy as np # for getting our distribution
import matplotlib.pyplot as plt # for plotting
import seaborn as sns; sns.set() # For a different plotting theme

# Don't worry so much about what rho is doing here
# rho of 1 will have a perfect upward slope
# rho of -1 will have a perfect downward slope
# rho of 0 will get us a cloud of points
r = 1

# Don't worry so much about the following three lines of code for now
# this is just getting the data for us to plot
mean = [15, 5]
cov = [[1, r], [r, 1]]
a, b = np.random.multivariate_normal(mean, cov, 400).T

# Adjust the figure size
plt.figure(figsize=(10,5))

# Plot the histograms of X and Y next to each other
plt.subplot(1,2,1)
plt.hist(x = x, bins = 15)
plt.title("X")

plt.subplot(1,2,2)
plt.hist(x = y, bins = 15)
plt.title("Y")

plt.show()
```
